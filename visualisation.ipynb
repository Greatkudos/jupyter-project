```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('survey_data.csv')
sns.histplot(df['Age'], bins=30, kde=True)
plt.title('Distribution of Age')
plt.show()
```

```python
sns.scatterplot(x='Age', y='Income', data=df)
plt.title('Relationship of Age to Income')
plt.show()
```

```python
sns.boxplot(x='MaritalStatus', y='Age', data=df)
plt.title('Age across Marital Status')
plt.show()
```

```python
corrolation_matrix = df.corr(numeric_only=True)
sns.heatmap(corrolation_matrix, annot=True, cmap='coolwarm')
plt.title='Corrolation Matrix'
plt.show()
```

```python
sns.pairplot(df)
plt.show()
```
